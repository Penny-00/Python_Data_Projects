## What are the most demanded skills for the top 3 most popular data roles?

### Methodology

1. Clean up skill column
2. Calculate skill count based on job_title_short
3. Calculate Skill percentage
4. Plot final findings 

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_US = df[df['job_country'] == 'United States']
```

```python
# Exploding the job_skills column to count individual skills
df_skills = df_US.explode('job_skills')

df_skills_count = df_skills.groupby(['job_skills', 'job_title_short']).size().reset_index(name='skill_count')
df_skills_count.sort_values(by='skill_count', ascending=False, inplace=True)

df_skills_count
```

```python
job_titles = df_skills_count['job_title_short'].unique().tolist()

job_titles = sorted(job_titles[:3])  # Limiting to the first 3 job titles

job_titles
```

```python
fig, ax = plt.subplots(len(job_titles), 1)

for i, job_title in enumerate(job_titles):
    df_plot = df_skills_count[df_skills_count['job_title_short'] == job_title].head(5)
    df_plot.plot(kind='barh', x='job_skills', y='skill_count', ax=ax[i], title=job_title)
    ax[i].set_ylabel('')
    ax[i].invert_yaxis()  # Invert y-axis for better readability
    ax[i].legend().set_visible(False)  # Hide legend for clarity

fig.suptitle('Counts of Top Skills in Job Postings', fontsize=16)
plt.tight_layout() 
plt.show()
```

```python
df_job_title_count = df_US['job_title_short'].value_counts().reset_index(name='jobs_total') # Total count for each Job title we are interested in the US 
df_job_title_count
```

**Calculating Skills Percentages**

```python
pd.merge(df_skills_count, df_job_title_count, on='job_title_short', how='left') 
df_skills_perc = pd.merge(df_skills_count, df_job_title_count, on='job_title_short', how='left') 

df_skills_perc['skill_percent'] = (df_skills_perc['skill_count'] / df_skills_perc['jobs_total']) * 100 # Calculating the percentage of each skill in relation to the total number of job postings for that title

df_skills_perc
```

**Plotting Findings**

```python
fig, ax = plt.subplots(len(job_titles), 1)

for i, job_title in enumerate(job_titles):
    df_plot = df_skills_perc[df_skills_perc['job_title_short'] == job_title].head(5)
    sns.barplot(data=df_plot, x='skill_percent', y='job_skills', ax=ax[i], hue='skill_count', palette='dark:b_r')

    ax[i].set_title(job_title)
    ax[i].set_xlabel('')
    ax[i].set_ylabel('')
    ax[i].set_xlim(0, 80)  # Adjust x-axis limit for better visibility
    ax[i].legend().set_visible(False)  # Hide legend for clarity

    for n, v in enumerate(df_plot['skill_percent']):
        ax[i].text(v + 1, n, f"{v:.0f}%", color='black', va='center')

    if i != len(job_titles) - 1:
        ax[i].set_xticks([])  # Hide x-ticks for all but the last plot

fig.suptitle('Likelihood of Skills Requested in Job Postings', fontsize=16)
plt.tight_layout() 
plt.show()
```
