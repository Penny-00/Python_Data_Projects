```python
#Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df[pd.notna(df['salary_year_avg'])]['salary_year_avg']
```

```python
df_salary = df[pd.notna(df['salary_year_avg'])].copy()

def projected_salary(salary):
    return salary * 1.03

df_salary['salary_year_inflated'] = df_salary['salary_year_avg'].apply(projected_salary)

df_salary[['salary_year_avg', 'salary_year_inflated']]
```

```python
df_salary['salary_year_inflated'] = df_salary['salary_year_avg'].apply(lambda x: x * 1.03) # This is a more concise way to apply the function, an alternative to defining a separate function.

df_salary[['salary_year_avg', 'salary_year_inflated']]
```

```python
df_salary['salary_year_inflated'] = df_salary['salary_year_avg'] * 1.03 # This is another concise way to apply the function, an alternative to defining a separate function.

df_salary[['salary_year_avg', 'salary_year_inflated']]
```

```python
def projected_salary_adj(row):
    if "senior" in row['job_title_short']:
        return 1.05 * row['salary_year_avg']
    else:
        return 1.03 * row['salary_year_avg']


df_salary['salary_year_inflated_snr_adj'] = df_salary.apply(projected_salary_adj, axis=1) # This applies the function across the entire DataFrame row-wise.

```
